ISCA 2012 Conference paper scraping
---
Before starting, these are the links which the data will be scraping, and necessary libraries.

```python
from bs4 import BeautifulSoup
from dateutil import parser
import pandas as pd
import numpy as np
import requests

links = ["https://ieeexplore.ieee.org/xpl/mostRecentIssue.jsp?punumber=6230820&filter=issueId%20EQ%20%226236993%22&pageNumber=1",
         "https://ieeexplore.ieee.org/xpl/mostRecentIssue.jsp?punumber=6230820&punumber=6230820&filter=issueId%20EQ%20%226236993%22&pageNumber=2",
         "https://ieeexplore.ieee.org/xpl/mostRecentIssue.jsp?punumber=6230820&punumber=6230820&filter=issueId%20EQ%20%226236993%22&pageNumber=2&pageNumber=3"]
```

Let's scrape those pages...
---

```python
pages = []
for x in links:
    print("Processing page...")
    html = requests.get(str(x)).text
    pages.append(html.split("\n"))
```

```python
print(len(pages[0]))
```

Now that we've got all the pages, let's scrape the abstracts.
---

```python
hrefs = []
for page in pages:
    for line in page:
        line = line.encode("utf-8").decode("utf-8")
        if "View full abstract" in line:
            hrefs.append(line)
```

Let's see what we got, and clean it up into the links to the full abstract pages
---

```python
print(hrefs[0])
clean_refs = []
for item in hrefs:
    item = item.split("href=\"")[-1]
    item = item.split("\"> View")[0]
    clean_refs.append("https://ieeexplore.ieee.org" + str(item))
    
print(clean_refs[0])
```

Now that we've scraped the links to the full abstracts, let's scrape those pages, and retrieve the abstracts themselves.
---

```python
abstract_pages = []
for x in clean_refs:
    print("Processing page...")
    html = requests.get(str(x)).text
    abstract_pages.append(html)
```

```python
clean_abstracts = []
for page in abstract_pages:
    temp = page.split("\"abstract\":\"")
    abstract = temp[1]
    abstract = abstract.split("\",\"")[0]
    clean_abstracts.append(abstract)
```

```python
for x in clean_abstracts:
    print(x)
    print("**********************************************************************")
```
